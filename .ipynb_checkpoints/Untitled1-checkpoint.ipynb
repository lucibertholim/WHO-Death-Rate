```python
import pandas as pd
```

```python
df=pd.read_csv('Resources/HDI_countries.csv', encoding ='ISO-8859-1', skiprows=1)
```

```python
df.head()
```

```python
country=df['Country'].str.lstrip()
df['Country']=country
```

```python
df2=df[['Country','2000',
 '2001',
 '2002',
 '2003',
 '2004',
 '2005',
 '2006',
 '2007',
 '2008',
 '2009',
 '2010',
 '2011',
 '2012',
 '2013',
 '2014',
 '2015',
 '2016']]
```

```python
list(df2)
```

```python
df3=pd.melt(df2, id_vars=['Country'], value_vars=['2000',
 '2001',
 '2002',
 '2003',
 '2004',
 '2005',
 '2006',
 '2007',
 '2008',
 '2009',
 '2010',
 '2011',
 '2012',
 '2013',
 '2014',
 '2015',
 '2016'])
```
