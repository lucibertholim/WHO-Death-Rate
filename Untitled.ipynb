```python
import pandas as pd
```

```python
df_2000=pd.read_csv("ASDR2000_clean.csv")
df_2010=pd.read_csv("ASDR2010_clean.csv")
df_2016=pd.read_csv("ASDR2000_clean.csv")
```

```python
df_2000.columns = df_2000.columns.map(lambda x: str(x) + '_2000' if (x!= "6") else "Country")
df_2010.columns = df_2010.columns.map(lambda x: str(x) + '_2010' if (x!= "6") else "Country")
df_2016.columns = df_2016.columns.map(lambda x: str(x) + '_2016' if (x!= "6") else "Country")
```

```python
df_2000.head()
```

```python
df_merge=df_2000.merge(df_2010, how='left', on='Country', suffixes=(False, False))
```

```python
df_merge_final=df_merge.merge(df_2016, how='left', on='Country', suffixes=(False, False))
```
