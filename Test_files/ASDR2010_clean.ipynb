```python
import pandas as pd
```

```python
xls = pd.ExcelFile('GHE2016_Death-Rates-country.xls')
```

```python
df2010 = pd.read_excel(xls, 'ASDR2010', index_col=None, header=None)
```

```python
df2010.head(20)
```

```python
df2010_1=df2010.drop(df2010.index[0:6])
df2010_1.head()
```

```python
import numpy as np
df2010_1 = df2010_1.replace(np.nan, '', regex=True)
df2010_1.head(20)
```

```python
df2010_1["disease"] = df2010_1[3].map(str) + " " + df2010_1[4].map(str) + " " + df2010_1[5].map(str) + " " + df2010_1[6].map(str)   
```

```python
df2010_1.head(20)
```

```python
df2010_1b=df2010_1.drop([1,2,3,4,5,6], axis = 1) 
```

```python
#df2010_loc2=df2010_loc1.loc
df2010_1b.columns=df2010_1b.iloc[0]
df2010_1b.head(20)
```

```python
df2010_2=df2010_1b.drop(df2010_1b.index[0])
df2010_2.head(20)
```

```python
df2010_3=df2010_2.loc[df2010_2['Sex']=="Persons"]
df2010_3
```

```python
df2010_4=df2010_3.drop(['Sex'], axis=1)
```

```python
df2010_transposed=df2010_4.set_index('GHE cause   Member State\n(See Notes for explanation of colour codes)').T
```
