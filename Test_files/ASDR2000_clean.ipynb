```python
import pandas as pd
```

```python
xls = pd.ExcelFile('GHE2016_Death-Rates-country.xls')
```

```python
df2000 = pd.read_excel(xls, 'ASDR2000', index_col=None, header=None)
```

```python
df2000.head(20)
```

```python
df2000_1=df2000.drop(df2000.index[0:6])
df2000_1.head()
```

```python
import numpy as np
df2000_1 = df2000_1.replace(np.nan, '', regex=True)
df2000_1.head(20)
```

```python
df2000_1["disease"] = df2000_1[3].map(str) + " " + df2000_1[4].map(str) + " " + df2000_1[5].map(str) + " " + df2000_1[6].map(str)   
```

```python
df2000_1.head(20)
```

```python
df2000_1b=df2000_1.drop([1,2,3,4,5,6], axis = 1) 
```

```python
#df2000_loc2=df2000_loc1.loc
df2000_1b.columns=df2000_1b.iloc[0]
df2000_1b.head(20)
```

```python
df2000_2=df2000_1b.drop(df2000_1b.index[0])
df2000_2.head(20)
```

```python
df2000_3=df2000_2.loc[df2000_2['Sex']=="Persons"]
df2000_3
```

```python
df2000_4=df2000_3.drop(['Sex'], axis=1)
```

```python
df2000_transposed=df2000_4.set_index('GHE cause   Member State\n(See Notes for explanation of colour codes)').T
```
